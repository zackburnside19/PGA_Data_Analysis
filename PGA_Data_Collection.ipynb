```python
%%time
#package imports
from bs4 import BeautifulSoup
import lxml
import requests as request
import pandas as pd
import numpy as np
#access the url and assign
years = list(np.arange(2010, 2022))
dfs = []
for year in years:
    url = f'https://www.pgatour.com/content/pgatour/stats/stat.159.y{year}.html'
    request.get(url)
    page = request.get(url)

    #use bs4 
    soup = BeautifulSoup(page.text, 'html')

    #find the table
    table_data = soup.find('table', class_ ='table-styled')

    #assign the headers
    headers = []
    for i in table_data.find_all('th'): #find the positions of the headers
        title = i.text.strip() #convert to text
        headers.append(title)

    df = pd.DataFrame(columns = headers)

    #grab the contents of the table
    for j in table_data.find_all('tr')[1:]:
            row_data = j.find_all('td')
            row = [tr.text.strip() for tr in row_data]
#             length = len(df_2021)
            df.loc[len(df)] = row
    
    df['YEAR'] = year
    df['ROUNDS'] = df['ROUNDS'].astype(int)
    df['DISTANCE'] = df['DISTANCE'].astype(int)
    dfs.append(df)
final=pd.concat(dfs, axis=0)
```
